```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['svg.fonttype'] = 'none'
```

```python
hepg2_scores = [0.7702, 0.7700, 0.6706, 0.5058]
a549_scores = [0.7942, 0.7920, 0.6124, 0.5139]
gm12878_scores = [0.7872, 0.7886, 0.5779, 0.3989]
k562_scores = [0.7681, 0.7672, 0.6216, 0.4574]
tfs = ['CTCFf', 'CTCFr', 'YY1', 'SP1']
score_list = []
cell_line_list = []
tf_list = []
for i in range(len(hepg2_scores)):
    score_list.append(hepg2_scores[i])
    score_list.append(a549_scores[i])
    score_list.append(gm12878_scores[i])
    score_list.append(k562_scores[i])
    cell_line_list.append('HepG2')
    cell_line_list.append('A549')
    cell_line_list.append('GM12878')
    cell_line_list.append('K562')
    tf_list.append(tfs[i])
    tf_list.append(tfs[i])
    tf_list.append(tfs[i])
    tf_list.append(tfs[i])
data = {'score': score_list, 'cell_line': cell_line_list, 'tf': tf_list}
df = pd.DataFrame(data)
df
```

```python
# Generate a barplot of scores for each cell line
fig = plt.figure(figsize=(4, 3), dpi=200)
# sns.barplot(x='tf', y='score', data=df, hue='cell_line')
sns.barplot(x='tf', y='score', data=df, ci=100, palette=['#5299D8', '#5299D8', '#DC3220', '#5D3A9B'])
plt.title('First stage performance on unseen cell lines')
plt.ylabel("Pearson's R")
plt.xlabel('Transcription Factor')
# plt.legend(title='Cell Line')
for i in range(len(hepg2_scores)):
    mean_score = np.mean([hepg2_scores[i], a549_scores[i], gm12878_scores[i], k562_scores[i]])
    plt.text(i, mean_score+0.05, round(mean_score, 3), ha='center', va='bottom', fontsize=8)
plt.ylim(0, 1)
plt.tight_layout()
plt.savefig('./plots/first_stage_performance.svg')
```

```python
# Generate barplot of scores for each TF in HepG2
fig = plt.figure(figsize=(4, 3), dpi=200)
sns.barplot(x='tf', y='score', data=df[df['cell_line'] == 'HepG2'], palette=['#5299D8', '#5299D8', '#DC3220', '#5D3A9B'])
plt.title('First stage performance on HepG2')
plt.ylabel("Pearson's R")
plt.xlabel('Transcription Factor')
plt.ylim(0, 1)
# Add the actual scores to the plot
for i in range(len(hepg2_scores)):
    plt.text(i, hepg2_scores[i], round(hepg2_scores[i], 3), ha='center', va='bottom', fontsize=8)
plt.show()
```
